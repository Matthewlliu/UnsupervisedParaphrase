```python
import json
import os
from collections import defaultdict
```

```python
file_name = './book_categories.jsonl'
data = []
with open(file_name, 'r') as f:
    for line in f:
        data.append(json.loads(line))
```

```python
def split_cate(string):
    tmp = string.split(">>")
    tmp = [ t.strip() for t in tmp]
    return tmp
```

```python
def isEqual(cates):
    flag = True
    for i in range(len(cates)):
        if cates[0] != cates[i]:
            flag = False
            break
    return flag
```

```python
genre = defaultdict(int)
for entry in data:
    cates = entry['categories']
    if len(cates) == 1:
        parts = split_cate(cates[0])
        genre[parts[1]] += 1
    else:
        tmp = []
        for cate in cates:
            parts = split_cate(cate)
            tmp.append(parts[1])
        if isEqual(tmp):
            genre[tmp[0]] += 1
        else:
            #print(cates)
            for t in tmp:
                genre[t] += 1
```

```python
import numpy as np
np.sum([v for v in genre.values()])
```
